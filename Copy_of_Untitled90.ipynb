## **Analysis DataFrame (Column Names,Total Rows and Columns,Data Types of Columns,NUll values)**

```python
airbnb_df1 = airbnb_df.replace('[+,$]','', regex = True)                            # Removed unnecessary characters and save as aribnb_df1
airbnb_df1
```

```python
print("Total number of Rows and Columns in airbnb_df1 are:- ", airbnb_df1.shape)                       # Shape Function give the details of (No of Rows, No of Columns)
```

```python
def airbnb_info():
  temp = pd.DataFrame(index = airbnb_df1.columns)
  temp['Data Type'] = airbnb_df1.dtypes
  temp['Non-Null Values'] = airbnb_df1.count()
  temp['Unique Values'] = airbnb_df1.nunique()
  temp['NaN Values'] = airbnb_df1.isnull().sum()
  temp['NaN Values Percentage'] = (temp['NaN Values']/len(airbnb_df1))*100
  
  return temp
airbnb_info()
```

## **Data Filtering and Each Columns Data Operation**

### Here we have 4 columns('name','host_name','last_reviews','reviews_per_month') which contains some NaN values. we are not going to remove them from the table but are replacing them by placing 0 instead of "NaN"

```python
airbnb_df1['name'].fillna('0',inplace = True)                       # 'host_id has 16 NaN values

airbnb_df1['host_name'].fillna('0',inplace = True)                     # 'host_name' has 21 NaN values

airbnb_df1['last_review'].fillna('0',inplace = True)                   # 'last_review' has 10052 values

airbnb_df1['reviews_per_month'].fillna('0',inplace = True)              # 'reviews_per_month' has 10052 Values
```
